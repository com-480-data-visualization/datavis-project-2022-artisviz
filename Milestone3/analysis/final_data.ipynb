```python
import pandas as pd
import numpy as np
```

```python
# In the EU
avg_wg_EU = pd.read_csv('../data/avg_wg_EU.csv')
hrs_wkd_EU = pd.read_csv('../data/hrs_wkd_reduced_EU.csv')
gdp_hrs_wkd_EU = pd.read_csv('../data/gdp_hrs_wkd_reduced_EU.csv')
emp_rate_EU = pd.read_csv('../data/emp_rate_reduced_EU.csv')
inc_ineq_EU = pd.read_csv('../data/inc_ineq_reduced_EU.csv')
pvty_EU = pd.read_csv('../data/pvty_reduced_TOT_EU.csv')


#not in the EU
avg_wg_NEU = pd.read_csv('../data/avg_wg_NEU.csv')
pvty_NEU = pd.read_csv('../data/pvty_reduced_TOT_NEU.csv')
hrs_wkd_NEU = pd.read_csv('../data/hrs_wkd_reduced_NEU.csv')
gdp_hrs_wkd_NEU = pd.read_csv('../data/gdp_hrs_wkd_reduced_NEU.csv')
inc_ineq_NEU = pd.read_csv('../data/inc_ineq_reduced_NEU.csv')
emp_rate_NEU = pd.read_csv('../data/emp_rate_reduced_NEU.csv')
```

```python
df_EU = pd.merge(pvty_EU, gdp_hrs_wkd_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, inc_ineq_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, emp_rate_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, hrs_wkd_EU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, avg_wg_EU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
```

```python
df_EU
```

```python
df_NEU = pd.merge(pvty_NEU, gdp_hrs_wkd_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, inc_ineq_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, emp_rate_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, hrs_wkd_NEU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, avg_wg_NEU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
```

```python
df_NEU
```

```python
df_EU.to_csv('../data/final_EU.csv',na_rep='NaN')
df_NEU.to_csv('../data/final_NEU.csv',na_rep='NaN')
```
