```python
import pandas as pd
import numpy as np
```

```python
pvty = pd.read_csv('data/pvty.csv')
gdp_hrs_wkd = pd.read_csv('data/gdp_hrs_wkd.csv')
inc_ineq = pd.read_csv('data/inc_ineq.csv')
emp_rate = pd.read_csv('data/emp_rate.csv')
```

```python
inc_ineq
```

```python
test = pd.merge(pvty, gdp_hrs_wkd, how='outer', left_on=['LOCATION', 'Year'], right_on = ['LOCATION','Year'])
test = pd.merge(test, inc_ineq, how='outer', left_on=['LOCATION', 'Year'], right_on = ['LOCATION','Year'])
test = pd.merge(test, emp_rate, how='outer', left_on=['LOCATION', 'Year'], right_on = ['LOCATION','Year'])
```

```python
test
```

```python
test.to_csv('data/final.csv',na_rep='NaN')
```
