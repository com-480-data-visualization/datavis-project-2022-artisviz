```python
import pandas as pd
import numpy as np
```

```python
# In the EU
avg_wg_EU = pd.read_csv('../data/avg_wg_EU.csv')
hrs_wkd_EU = pd.read_csv('../data/hrs_wkd_reduced_EU.csv')
gdp_hrs_wkd_EU = pd.read_csv('../data/gdp_hrs_wkd_reduced_EU.csv')
emp_rate_EU = pd.read_csv('../data/emp_rate_reduced_EU.csv')
inc_ineq_EU = pd.read_csv('../data/inc_ineq_reduced_EU.csv')
pvty_EU = pd.read_csv('../data/pvty_reduced_TOT_EU.csv')


#not in the EU
avg_wg_NEU = pd.read_csv('../data/avg_wg_NEU.csv')
pvty_NEU = pd.read_csv('../data/pvty_reduced_TOT_NEU.csv')
hrs_wkd_NEU = pd.read_csv('../data/hrs_wkd_reduced_NEU.csv')
gdp_hrs_wkd_NEU = pd.read_csv('../data/gdp_hrs_wkd_reduced_NEU.csv')
inc_ineq_NEU = pd.read_csv('../data/inc_ineq_reduced_NEU.csv')
emp_rate_NEU = pd.read_csv('../data/emp_rate_reduced_NEU.csv')
```

```python
df_EU = pd.merge(pvty_EU, gdp_hrs_wkd_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, inc_ineq_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, emp_rate_EU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, hrs_wkd_EU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_EU = pd.merge(df_EU, avg_wg_EU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
```

```python
df_EU
```

```python
df_NEU = pd.merge(pvty_NEU, gdp_hrs_wkd_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, inc_ineq_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, emp_rate_NEU, how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, hrs_wkd_NEU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
df_NEU = pd.merge(df_NEU, avg_wg_NEU , how='outer', left_on=['country', 'Year'], right_on = ['country','Year'])
```

```python
df_NEU
```

```python
df_EU.to_csv('../data/final_EU.csv',na_rep='NaN')
df_NEU.to_csv('../data/final_NEU.csv',na_rep='NaN')
```

# Generate the csv data for the correlation matrix

```python
df_EU_corr = df_EU.drop(['Year','country'], axis = 1)
df_NEU_corr = df_NEU.drop(['Year','country'], axis = 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def dataframe_description(dataframe):
    print('Some statistics on the dataframe \n')
    print(dataframe.describe())
    print('There are Nan values in this dataframe:', dataframe.isnull().values.any())
    for name in dataframe.keys():
            print(name, 'has', dataframe[name].isnull().sum(), 'Nan values')

    print('The dataframe has', len(dataframe[dataframe.duplicated(keep=False)]), 'duplicated values')    
    print('The correlation matrix is the following\n')
    corr_mat = dataframe.corr()
    fig, ax = plt.subplots(figsize = (6,4)) # create the figure object
    sns.heatmap(corr_mat, vmax = 1, square = True) # plot the heat-map
    return corr_mat
```

```python
corr_EU = df_EU_corr.corr().round(2)
corr_NEU = df_NEU_corr.corr().round(2)

corr_EU = corr_EU.stack().reset_index()
corr_EU.columns = ['Field 1','Field 2','Value']
corr_NEU = corr_NEU.stack().reset_index()
corr_NEU.columns = ['Field 1','Field 2','Value']
```

```python
corr_EU.to_csv('../data/correlation_EU.csv')
corr_NEU.to_csv('../data/correlation_NEU.csv')
corr_EU.to_json('../data/correlation_EU.json')
corr_NEU.to_json('../data/correlation_NEU.json')
```

Checking which correlations are the most different between EU countries and non EU countries

```python
fig, ax = plt.subplots(figsize = (6,4)) # create the figure object
sns.heatmap(corr_NEU - corr_EU, vmax = 1, square = True) # plot the heat-map
```

# Correlation per country

```python
EU = ['AUT', 'BEL', 'CZE', 'DNK', 'FIN', 'FRA', 'DEU',
       'GRC', 'HUN', 'IRL', 'ITA', 'LUX', 'NLD', 'POL', 'PRT', 'SVK', 'ESP', 'SWE','EST', 'SVN', 'LVA', 'LTU'] #no croatia, bulgaria, romania, malta, cyprus

notEU = ['AUS','CAN','KOR','MEX','NOR','CHE','GBR','USA','ISR','ISL','NZL','CHL']
```

```python
df_EU_corr = df_EU.drop(['Year'], axis = 1)
df_NEU_corr = df_NEU.drop(['Year'], axis = 1)
total_corr = pd.DataFrame()
for country in EU:
    temp_df = df_EU_corr[df_EU_corr["country"] == country ]
    temp_corr = temp_df.corr().round(2)
    temp_corr = temp_corr.stack().reset_index()
    temp_corr["country"]=country
    temp_corr.columns = ['Field 1','Field 2','Value', 'country']
    total_corr = pd.concat([total_corr,temp_corr])
total_corr.to_csv('../data/correlation_EU_each.csv') 
```

```python
df_NEU_corr = df_NEU.drop(['Year'], axis = 1)
total_corr = pd.DataFrame()
for country in notEU:
    temp_df = df_NEU_corr[df_NEU_corr["country"] == country ]
    temp_corr = temp_df.corr().round(2)
    temp_corr = temp_corr.stack().reset_index()
    temp_corr["country"]=country
    temp_corr.columns = ['Field 1','Field 2','Value', 'country']
    total_corr = pd.concat([total_corr,temp_corr])

total_corr.to_csv('../data/correlation_NEU_each.csv')
```
