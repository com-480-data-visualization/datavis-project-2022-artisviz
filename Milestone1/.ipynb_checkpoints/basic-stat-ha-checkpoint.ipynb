**Évolution au cours du temps de l'une des variables**

```python
#Inclusions
import pandas as pd
import numpy as np
```

```python
hrs_wk = pd.read_csv('hours_worked.csv')
```

```python
hrs_wk.head()
```

```python
hrs_CHE = hrs_wk[hrs_wk['LOCATION']=='CHE']
```

```python
hrs_CHE.head()
```

```python
import seaborn as sns
```

```python
sns.jointplot(data=hrs_CHE, x="TIME", y="Value", kind='reg', marginal_kws=dict(bins=25))
```

```python
hrs_CHE.corr(method='pearson')
```

```python
#évolution au cours du temps des average wages en Suisse:
avg_wg = pd.read_csv('average_wages.csv')
avg_wg_CHE = avg_wg[avg_wg['LOCATION']=='CHE']
sns.jointplot(data=avg_wg_CHE, x="TIME", y="Value", kind='reg', marginal_kws=dict(bins=25))
```

```python
avg_wg_CHE.corr(method='pearson')
```

```python
avg_wg_CHE.plot(x = 'TIME', y = 'Value')
```

```python
#évolution au cours du temps des average wages en Suisse:
hrs_pvty = pd.read_csv('workHours_exitPoverty.csv')
hrs_pvty_CHE = hrs_pvty[hrs_pvty['LOCATION']=='CHE']
sns.jointplot(data=hrs_pvty_CHE, x="TIME", y="Value", kind='reg', marginal_kws=dict(bins=25))
```
